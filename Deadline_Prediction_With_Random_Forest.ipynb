```python
# importing modules
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
```

```python
# Loading the data
final_df = pd.read_csv('final_df.csv')
```

```python
final_df.head()
```

```python
final_df.describe()
```

```python
df = final_df.drop(['Unnamed: 0', 'id', 'task_id', 'number_of_files', 'deadline_set_at', 'deadline', 'deadline_closed_at',
                    'actual task duration', 'planned task duration'], axis=1)
```

```python
df.head()
```

```python
df_onehot = pd.get_dummies(df)
```

```python
df_onehot.shape
```

```python
X = df_onehot.drop(['actual task duration in minutes'], axis=1)
Y = df_onehot['actual task duration in minutes']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
x_train_final = x_train.copy()
x_test_final = x_test.copy()
```

```python
scaler = StandardScaler()
x_train_final[['action_id', 'from_status_id', 'deadline_set_on', 'workflow_id', 'to_status_id',
               'deadline_set_by', 'planned task duration in minutes']] = scaler.fit_transform(x_train_final[['action_id', 'from_status_id',
               'deadline_set_on', 'workflow_id', 'to_status_id','deadline_set_by', 'planned task duration in minutes']])
```

```python
x_test_final[['action_id', 'from_status_id', 'deadline_set_on', 'workflow_id', 'to_status_id',
              'deadline_set_by', 'planned task duration in minutes']] = scaler.transform(x_test_final[['action_id', 'from_status_id',
              'deadline_set_on', 'workflow_id', 'to_status_id', 'deadline_set_by', 'planned task duration in minutes']])
```

```python
model = RandomForestRegressor(random_state=42)
model.fit(x_train_final, y_train)
```

```python
mean_squared_error(y_test, model.predict(x_test_final), squared=False)
```

```python
model.score(x_test_final, y_test)
```

```python
model.score(x_train_final, y_train)
```
